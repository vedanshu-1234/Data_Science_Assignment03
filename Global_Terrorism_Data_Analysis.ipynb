## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
import pandas as pd

# Load dataset
df = pd.read_csv("Global Terrorism Data.csv", encoding='latin1')

# Check basic info
print(df.shape)
print(df.info())

# Check missing values
print(df.isnull().sum())

# Drop columns with more than 50% missing values
df = df.dropna(axis=1, thresh=len(df)*0.5)

# Fill remaining missing values
df.fillna("Unknown", inplace=True)

# Check duplicates
print(df.duplicated().sum())

# Check unique values
print(df.nunique())
````

#### Chart - 13

````python
# Chart - 13 visualization code
# Clean data
nkill = pd.to_numeric(df['nkill'], errors='coerce')
nwound = pd.to_numeric(df['nwound'], errors='coerce')
success = pd.to_numeric(df['success'], errors='coerce')

# Calculate means (ignore NaN)
values = [
    nkill.mean(skipna=True),
    nwound.mean(skipna=True),
    success.mean(skipna=True)
]

labels = ['Kills', 'Wounds', 'Success']

# Convert to numpy array
values = np.array(values, dtype=float)

# Create angles
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

# Close the circle
values = np.concatenate((values, [values[0]]))
angles = np.concatenate((angles, [angles[0]]))

# Plot
plt.figure(figsize=(6,6))
plt.polar(angles, values)
plt.fill(angles, values, alpha=0.3)

plt.xticks(angles[:-1], labels)
plt.title("Radar Chart of Attack Features")
plt.show()
````

# **Applying 5 ML Technique**

````python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Global Terrorism Data.csv", encoding='latin1')

# Show first rows
print(df.head())

# Select important columns (drop missing values)
df = df[['nkill', 'nwound', 'success', 'iyear', 'country_txt']]
df = df.dropna()
````

## **Regression (Predict Number of Kills)**

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[['nwound', 'iyear']]
y = df['nkill']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)

print("Sample Predictions:", pred[:5])
````

## **Classification (Attack Success Prediction)**

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = df[['nkill', 'nwound', 'iyear']]
y = df['success']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pred))
````

## **Clustering (Group Similar Attacks)**

````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = df[['nkill', 'nwound']]

kmeans = KMeans(n_clusters=3)
df['cluster'] = kmeans.fit_predict(X)

plt.scatter(df['nkill'], df['nwound'], c=df['cluster'])
plt.xlabel("Kills")
plt.ylabel("Wounds")
plt.title("Clustering of Attacks")
plt.show()
````

## **Time Series Analysis (Attacks per Year)**

````python
yearly = df.groupby('iyear').size()

yearly.plot()
plt.title("Attacks Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
````

## **Anomaly Detection (Unusual Attacks)**

````python
from sklearn.ensemble import IsolationForest

X = df[['nkill', 'nwound']]

model = IsolationForest(contamination=0.01)
df['anomaly'] = model.fit_predict(X)

print(df[df['anomaly'] == -1].head())  # Show anomalies
````
